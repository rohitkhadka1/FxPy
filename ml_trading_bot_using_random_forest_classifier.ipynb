```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import yfinance as yf

```

```python
# Step 1: Get Historical Data
data = yf.download("EURUSD=X", start="2020-01-01", end="2024-12-31", interval="1d")
data.dropna(inplace=True)

# Step 2: Create Technical Indicators
data['SMA_10'] = data['Close'].rolling(window=10).mean()
data['SMA_50'] = data['Close'].rolling(window=50).mean()
data['RSI'] = 100 - (100 / (1 + data['Close'].pct_change().rolling(window=14).mean() / data['Close'].pct_change().rolling(window=14).std()))

# Step 3: Define Target Variable (1 if price goes up next day, else 0)
data['Target'] = np.where(data['Close'].shift(-1) > data['Close'], 1, 0)

# Step 4: Prepare Features and Labels
features = ['SMA_10', 'SMA_50', 'RSI']
data.dropna(inplace=True)
X = data[features]
y = data['Target']

# Step 5: Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

```

```python
# Step 6: Train Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Step 7: Evaluate
y_pred = model.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")


```

```python
# Step 8: Simulated Signal Output
data['Signal'] = model.predict(X)
data[['Close', 'Signal']].tail(10)
```
